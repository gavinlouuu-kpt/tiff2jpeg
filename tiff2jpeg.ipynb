```python
from PIL import Image

input_folder = 'Input/Test_images/2'
output_folder = 'Output/2_PNG'
original_format = '.tiff' # Include the dot

```

```python
import os
from PIL import Image

def convert_tiff_to_high_quality_png(input_folder, output_folder):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Loop through all the files in the input folder
    for file in os.listdir(input_folder):
        if file.lower().endswith(('.tiff', '.tif')):
            # Construct full file paths
            input_path = os.path.join(input_folder, file)
            output_filename = os.path.splitext(file)[0] + '.png'
            output_path = os.path.join(output_folder, output_filename)

            try:
                # Open the image
                with Image.open(input_path) as img:
                    # Save the image as highest quality PNG
                    img.save(output_path, 'PNG', compress_level=0)
                print(f"Converted: {file} -> {output_filename}")
            except Exception as e:
                print(f"Error converting {file}: {str(e)}")

# Usage


convert_tiff_to_high_quality_png(input_folder, output_folder)
```
